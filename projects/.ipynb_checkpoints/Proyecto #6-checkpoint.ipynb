# Description

You work for the online store Ice that sells video games all over the world. User and expert reviews, genres, platforms (e.g. Xbox or PlayStation), and historical data on game sales are available in open sources. You have to identify patterns that determine whether a game is successful or not. This will allow you to detect promising projects and plan advertising campaigns.

In front of you is data going back to 2016. Let's imagine it's December 2016 and you're planning a campaign for 2017.

The important thing is to gain experience working with data. It doesn't really matter whether you're forecasting 2017 sales based on 2016 data or 2027 sales based on 2026 data.

The dataset contains a "rating" column that stores the ESRB rating of each game. The Entertainment Software Rating Board evaluates the content of a game and assigns an age rating such as Teen or Mature.

```python
#import the libraries
import pandas as pd
import matplotlib.pyplot as plt
import streamlit as st
import seaborn as sns
from scipy import stats as st
```

```python
#Make column names lowercase
games_data = pd.read_csv('/datasets/games.csv')
games_data.columns = games_data.columns.str.lower()
display(games_data.head())
```

```python
#Checking for null values within the dataframe
missing_values = games_data.isnull().sum()
print(missing_values)
```

```python
display(games_data.info())
```

```python
#Convert data
columns_to_replace = ['year_of_release', 'critical_score']

# Convert 'year_of_release' to object data type
games_data['year_of_release'] = games_data['year_of_release'].astype(object)

#Convert 'critic_score' to object
games_data['critic_score'] = games_data['critic_score'].astype(object)


for column in columns_to_replace:
    games_data[column].fillna('unknown', inplace=True)


display(games_data.info())
```

```python
# Loop in headers replacing missing values with 'unknown'
columns_to_replace = ['year_of_release', 'critic_score', 'genre','name']

for column in columns_to_replace:
    games_data[column].fillna('unknown', inplace=True)
```

```python
missing_values = games_data.isnull().sum()
print(missing_values)
```

```python
games_data['total_sales'] = games_data['na_sales'] + games_data['eu_sales'] + games_data['jp_sales'] + games_data['other_sales']

# Show the DataFrame with the new total sales column
display(games_data.head())
```

### Step 3 Analyze the data ###

How many games were released in different years?
Are the data from each period significant?

```python
#Prevent it from taking unknown as a value
valid_years = games_data['year_of_release'] != 'unknown'
filtered_data = games_data[valid_years]

# Count the number of games released per year after filtering
games_per_year = filtered_data.groupby('year_of_release')['name'].count()

# Graph the number of games released per year
plt.figure(figsize=(12, 6))
plt.plot(games_per_year.index, games_per_year.values, marker='o')
plt.title('Number of games released per year')
plt.xlabel('Release year')
plt.ylabel('Number of Games')
plt.grid(True)
plt.show()
```

### Notice how sales vary from one platform to another?

```python
platform_sales = games_data.groupby('platform')[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum()

# Calcular las ventas totales por año para las plataformas con mayores ventas
top_platforms = platform_sales.sum(axis=1).nlargest(5).index
sales_by_year = games_data[games_data['platform'].isin(top_platforms)].groupby(['year_of_release', 'platform'])[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum().unstack()
top_platforms = platform_sales.sum(axis=1).nlargest(5).index
print(top_platforms)
```

# As sales vary from one platform to another?

```python
top_platforms = platform_sales.sum(axis=1).nlargest(5).index
top_platforms_to_plot = [str(platform) for platform in top_platforms]

# Filter data by selected platforms
filtered_sales_by_year = sales_by_year.loc[:, sales_by_year.columns.get_level_values('platform').isin(top_platforms_to_plot)]

years = filtered_sales_by_year.index.get_level_values('year_of_release')
plt.figure(figsize=(12, 6))
sns.kdeplot(data=filtered_sales_by_year, fill=True, common_norm=False, palette='husl')

# Map platform labels to real names
platform_names = {'PS2': 'PlayStation 2', 'X360': 'Xbox 360', 'PS3': 'PlayStation 3', 'Wii': 'Wii', 'DS': 'Nintendo DS'}
platform_labels = [platform_names.get(platform, platform) for platform in top_platforms_to_plot]

# Configure the legend with the platform names
plt.legend(title='Platform', labels=platform_labels, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Platform Sales Distribution')
plt.xlabel('Total Sales')
plt.ylabel('Distribution')
plt.show()
```

## Platforms that used to be popular but now have no sales.
## How long does it generally take for new platforms to appear and old ones to disappear?

```python
available_years = games_data['year_of_release'].dropna().unique()

# Obtener las consolas con mayores y menores ventas
top_platforms = games_data.groupby('platform')['total_sales'].sum().nlargest(3).index
bottom_platforms = games_data.groupby('platform')['total_sales'].sum().nsmallest(3).index

# Filtrar los datos para incluir solo las consolas con mayores y menores ventas
filtered_sales_by_year_top = games_data[games_data['platform'].isin(top_platforms)].pivot_table(index='year_of_release', columns='platform', values='total_sales', aggfunc='sum')
filtered_sales_by_year_bottom = games_data[games_data['platform'].isin(bottom_platforms)].pivot_table(index='year_of_release', columns='platform', values='total_sales', aggfunc='sum')

# Combinar las dos tablas filtradas
filtered_sales_by_year = pd.concat([filtered_sales_by_year_top, filtered_sales_by_year_bottom], axis=1)

plt.figure(figsize=(12, 6))
filtered_sales_by_year.plot(kind='bar', stacked=True)
plt.title('Total Sales by Year and Platform')
plt.xlabel('Release Year')
plt.ylabel('Total Sales')
plt.legend(title='Platform', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## How user and professional reviews affect sales of a popular platform.

```python
# Select the best selling platform
popular_platform = top_platforms[0]

# Filter the data for the best selling platform
platform_data = games_data[games_data['platform'] == popular_platform]

# Convert critical_score scores to numerical type
#platform_data['critic_score'] = pd.to_numeric(platform_data['critic_score'], errors='coerce')
platform_data.loc[:, 'critic_score'] = pd.to_numeric(platform_data['critic_score'], errors='coerce')

# Convert user ratings to numeric
#platform_data['user_score'] = pd.to_numeric(platform_data['user_score'], errors='coerce')
platform_data.loc[:, 'user_score'] = pd.to_numeric(platform_data['user_score'], errors='coerce')


plt.figure(figsize=(12, 6))
sns.scatterplot(x='critic_score', y='total_sales', data=platform_data, label='Critic Score', alpha=0.7)
sns.scatterplot(x='user_score', y='total_sales', data=platform_data, label='User Score', alpha=0.7)
plt.title(f'Relationship between Reviews and Sales for the platform {popular_platform}')
plt.xlabel('Score')
plt.ylabel('Total Sales')
plt.legend(title='Review Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Calculate the correlation coefficient
correlation_critic = platform_data['critic_score'].corr(platform_data['total_sales'])
correlation_user = platform_data['user_score'].corr(platform_data['total_sales'])

display(f"Correlation coefficient between critical_score and total_sales: {correlation_critic}")
display(f"Correlation coefficient between user_score and total_sales: {correlation_user}")
```

# Test the following hypotheses:
— Average user ratings for Xbox One and PC platforms are the same.

```python
# Filter data to get user ratings for Xbox One
xbox_one_user_scores = games_data[(games_data['platform'] == 'XOne') & (games_data['user_score'] != 'tbd')]['user_score'].dropna().astype(float)
# Filter data to get user ratings for PC
pc_user_scores = games_data[(games_data['platform'] == 'PC') & (games_data['user_score'] != 'tbd')]['user_score'].dropna().astype(float)
t_statistic, p_value = st.ttest_ind(xbox_one_user_scores, pc_user_scores, equal_var=False)

alpha = 0.05

print("Statistic t:", t_statistic)
print("P value:", p_value)

if p_value < alpha:
    #There is enough evidence to say that the average user ratings for Xbox One and PC are different.
    print("The null hypothesis is rejected")
else:
    #There is not enough evidence to say that the average user ratings for Xbox One and PC are different.
    print("The null hypothesis is not rejected")
```

— Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes.

```python
action_user_scores = games_data[(games_data['genre'] == 'Action') & (games_data['user_score'] != 'tbd')]['user_score'].dropna().astype(float)
sports_user_scores = games_data[(games_data['genre'] == 'Sports') & (games_data['user_score'] != 'tbd')]['user_score'].dropna().astype(float)

t_statistic, p_value = st.ttest_ind(action_user_scores, sports_user_scores, equal_var=False)

alpha = 0.05

print("Statistic t:", t_statistic)
print("P value:", p_value)

# Compare with significance level
if p_value < alpha:
    #There is enough evidence to say that average user ratings for action and sports genres are different
    print("The null hypothesis is rejected.")
else:
    #There is not enough evidence to say that average user ratings for action and sports genres are different
    print("The null hypothesis is not rejected")
    
```
