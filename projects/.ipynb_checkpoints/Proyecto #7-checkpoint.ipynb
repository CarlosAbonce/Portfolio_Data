```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# CSV file paths
file_path_01 = "/datasets/project_sql_result_01.csv"
file_path_04 = "/datasets/project_sql_result_04.csv"
file_path_07 = "/datasets/project_sql_result_07.csv"

# Import the CSV files
df_taxi = pd.read_csv(file_path_01)
df_barrio = pd.read_csv(file_path_04)
df_rain = pd.read_csv(file_path_07)
```

```python
# Show the first rows of each DataFrame to verify the import
print("First rows of project_sql_result_01.csv:")
print(df_taxi.head())

print("\nFirst rows of project_sql_result_04.csv:")
print(df_barrio.head())
```

```python
print(df_taxi.info())

print("\nFirst rows of project_sql_result_04.csv:")
print(df_barrio.info())
```

```python
# Identify the top 10 neighborhoods in terms of tour completion
df_sorted = df_barrio.sort_values(by='average_trips', ascending=False)
top_10_neighborhoods = df_sorted.head(10)
print(top_10_neighborhoods[['dropoff_location_name', 'average_trips']])
```

```python
# Graph of taxi companies and number of trips
plt.figure(figsize=(12, 6))
sns.barplot(x='company_name', y='trips_amount', data=df_taxi)
plt.title('Number of Trips per Taxi Company')
plt.xlabel('Taxi Company')
plt.ylabel('Number of Trips')
plt.xticks(rotation=90, ha='right')
plt.show()
```

# Taxi companies such as: Flash Cab, Taxi Affiliation Services and Medallion leasing are the most requested companies and with the highest number of trips.

```python
# Chart of the top 10 neighborhoods
plt.figure(figsize=(12, 6))
sns.barplot(x='dropoff_location_name', y='average_trips', data=df_barrio.sort_values(by='average_trips', ascending=False).head(10))
plt.title('Top 10 Neighborhoods for travel')
plt.xlabel('Neighborhood')
plt.ylabel('Average Trips')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Loop, River North and Streeterville are the neighborhoods where trips are requested the most

Based on the data in the graphs, I consider that the least requested companies such as RC Andrews Cab or Felman Corp should change their strategy or change the neighborhood where they are located, since unlike other taxi companies they do not have a very high demand.

```python
print(df_rain.head(20))
print(df_rain.info())
```

```python
df_rain['start_ts'] = pd.to_datetime(df_rain['start_ts'])

# Select only rainy Saturdays
saturdays_rainy = df_rain[(df_rain['weather_conditions'] == 'Bad') & (df_rain['start_ts'].dt.dayofweek == 5)]['duration_seconds']


alpha = 0.05
statistic, p_value = stats.ttest_1samp(saturdays_rainy, popmean=df_rain['duration_seconds'].mean())

if p_value < alpha:
    print("We reject the null hypothesis. \nThe average duration changes on rainy Saturdays.")
else:
    print("We cannot reject the null hypothesis. \nThe average duration changes on rainy Saturdays.")
```
